```python
%%writefile README.md

# Teoría y desarrollo práctico de un proyecto de detección de Neumonía con Convolutional Neural Networks

### Un modelo de IA para detectar neumonía en una radiografía sirve como una herramienta de apoyo crucial para los médicos, no como un reemplazo, ofreciendo beneficios significativos en eficiencia, precisión y acceso a la atención.

Las razones clave para su utilidad son:

  - Aumento de la precisión diagnóstica:
    - Aunque un médico experimentado puede identificar la neumonía sin problema, los algoritmos de IA pueden detectar anomalías sutiles o signos tempranos que podrían pasar desapercibidos para el ojo humano, especialmente en casos complejos o de difícil identificación, como fisuras finas u opacidades iniciales.
  - Eficiencia y priorización (Triage):
    - La IA puede analizar una imagen y generar un resultado preliminar en minutos o incluso segundos, mucho más rápido que la revisión humana. Esto permite priorizar los casos urgentes (como neumotórax o embolismo pulmonar) en grandes volúmenes de trabajo, asegurando que los pacientes más graves reciban atención más rápidamente.
  - Soporte a médicos menos experimentados:
    - En entornos con escasez de radiólogos o personal médico con menos experiencia (como servicios de urgencias nocturnos o centros de salud rurales), la IA proporciona un "segundo par de ojos" experto, aumentando la seguridad y fiabilidad del diagnóstico.
  - Reducción de la carga de trabajo y el agotamiento:
    - La IA puede encargarse de la revisión inicial de casos negativos o de baja probabilidad, liberando tiempo valioso para que los radiólogos se concentren en los casos más desafiantes o en otras tareas clínicas.
  - Resultados consistentes y reproducibles:
    - A diferencia de los humanos, cuyo rendimiento puede fluctuar debido a la fatiga o la variabilidad individual, los algoritmos de IA proporcionan resultados consistentes y objetivos cada vez.
  - Acceso a zonas remotas:
    - En áreas geográficas donde no hay acceso inmediato a un especialista en radiología, un sistema de IA puede analizar las imágenes localmente o a través de la nube, facilitando un diagnóstico oportuno.

En resumen, el modelo de IA es una herramienta de apoyo robusta que optimiza el flujo de trabajo, mejora la fiabilidad diagnóstica general y garantiza una atención más rápida y equitativa para los pacientes.





### Gracias a https://www.kaggle.com/code/kashyapgohil/pneumonia-detection-using-cnn/notebook
---

## Objetivos del proyecto:
  1. Teoría de las redes convolucionales.
  2. ***Principalmente***, revisión práctica de los principales librerías, funciones y comandos normalmente utilizados durante el desarrollo de un proyecto real, desde cero, de computer vision con una red CNN, especialmente Open CV (CV2).
  3. Explicación del ciclo de vida del desarrollo de un proyecto de Computer Vision para la detección de Neumonía.
  4. Demostración del uso de la librería OpenCV (CV2) para el preprocesamiento de las imágenes.
  5. Comprobación de la precisión de predicción.
  6. Ajuste de hiperparámetros para optimización del modelo.

  ---

## Redes convolucionales

Convolutional Neural Networks son sistemas de inteligencia artificial basados en redes neurales de varias capas que pueden identificar,reconocer y clasificar objetos, asi como detectar y segmentar objetos dentro de la imagen.

La operación de convolución combina dos funciones o datasets (la imagen y un filtro o "kernel") para producir un mapa de características ("features map").

Las capas convolucionales realizan una operación matemática llamada convolución, una especie de multiplicación matricial especializada, sobre los datos de entrada. La operación de convolución ayuda a preservar la relación espacial entre píxeles aprendiendo las características de la imagen usando pequeños cuadrados de datos de entrada.

El procesamiento de imágenes en una Red Neuronal Convolucional (CNN) funciona mediante una serie de capas que aprenden y extraen características automáticamente de los datos visuales, desde rasgos básicos como bordes hasta patrones complejos, sin necesidad de ingeniería manual de características.

El proceso paso a paso es el siguiente:
1. Capa de Entrada (Input Layer):
    - La imagen digital (que es una matriz de píxeles) se introduce en la red.
2. Capas Convolucionales (Convolutional Layers):
    - Esta es la capa principal y donde ocurre la mayor parte del procesamiento.
    - Se aplican filtros (también llamados núcleos o kernels) a la imagen. Cada filtro se desliza sobre la imagen (operación de convolución), detectando patrones locales específicos, como bordes, texturas o esquinas.
    - La salida de cada filtro es un mapa de características (feature map), que indica dónde se encontró ese patrón específico en la imagen.
    - A menudo, después de la convolución, se aplica una función de activación no lineal (como ReLU, Rectified Linear Unit) para introducir no linealidad en el modelo.
3. Capas de Agrupamiento (Pooling Layers):
    - Estas capas se intercalan entre las capas convolucionales.
    - Su función es reducir la dimensionalidad de los mapas de características, lo que disminuye la cantidad de cálculos y ayuda a evitar el sobreajuste (overfitting).
    - El tipo más común es el Max Pooling, que toma el valor máximo de un área pequeña del mapa de características, conservando la información más relevante.
4. Capas Totalmente Conectadas (Fully Connected Layers):
    - Después de varias combinaciones de capas convolucionales y de agrupamiento, los datos se "aplanan" (flatten) en un vector unidimensional.
    - Este vector se pasa a una red neuronal tradicional (totalmente conectada), que utiliza las características extraídas para realizar la tarea final de clasificación o detección.
    - La capa de salida proporciona el resultado final, por ejemplo, la probabilidad de que la imagen pertenezca a una clase determinada (un gato, un perro, un coche, etc.).

En resumen, las primeras capas de la CNN identifican elementos básicos y, a medida que la información pasa por capas más profundas, la red aprende a reconocer características cada vez más complejas y abstractas, hasta poder clasificar la imagen completa.

---
## Open CV

OpenCV es una biblioteca de código abierto para visión artificial que proporciona herramientas y algoritmos para procesamiento de imágenes, análisis de video y aprendizaje automático en tiempo real. Se utiliza para tareas como detección de rostros, reconocimiento de objetos y conducción autónoma, y es compatible con varios lenguajes como C++, Python y Java.
  - Características principales:
    - Código abierto y gratuito:
      - Publicada bajo la licencia BSD, lo que permite su uso tanto para fines comerciales como de investigación.
    - Multiplataforma:
      - Funciona en sistemas operativos como Linux, Windows, Mac OS y Android.
    - Multilenguaje:
      - Soporta múltiples lenguajes de programación, con C++ y Python como los más comunes, pero también Java y MATLAB.
    - Algoritmos optimizados:
      - Incluye más de 2.500 algoritmos optimizados para tareas de visión por computadora y aprendizaje automático en tiempo real.
    - Amplia gama de aplicaciones:
      - Se utiliza para proyectos como realidad aumentada, robótica, seguridad, análisis de movimiento y más.

¿Cómo funciona?

  - ***Un flujo de trabajo típico en OpenCV implica cargar una imagen o video, aplicar transformaciones (como filtrado o detección de características) y luego mostrar o guardar el resultado. Es una herramienta versátil que puede trabajar por sí sola o en combinación con otros marcos de aprendizaje profundo.***
---
## Ciclo de vida del desarrollo de un proyecto de Computer Vision

El ciclo de vida del desarrollo de un proyecto de Computer Vision es un proceso iterativo que adapta las fases tradicionales del desarrollo de software y del machine learning para abordar las particularidades de los datos visuales.
Las fases principales son:
  1. Definición y Planificación del Alcance:
      - Identificar el problema de negocio y el objetivo específico que la visión por computadora resolverá (por ejemplo, detección de defectos en una línea de producción, reconocimiento facial, etc.).
      - Establecer los requisitos del sistema, incluyendo la precisión esperada, la velocidad de procesamiento y las limitaciones de hardware/software.
      - Planificar la recopilación de datos, el presupuesto y el cronograma del proyecto.
  2. Recopilación y Preparación de Datos:
      - Adquisición de imágenes/videos: Recolectar un conjunto de datos visuales relevante y representativo del entorno real.
      - Limpieza y preprocesamiento: Normalizar, redimensionar y limpiar los datos para garantizar su calidad y uniformidad.
      - Anotación/Etiquetado: Etiquetar manualmente los datos (por ejemplo, delimitar objetos con cuadros, segmentar regiones) para que el modelo de aprendizaje supervisado pueda aprender.
      - Aumento de los datos: Aumentar artificialmente la variabilidad del conjunto de datos de entrenamiento existente para mejorar la robustez del modelo y reducir el sobreajuste (overfitting). Esto implica diseñar las transformaciones (rotaciones, recortes, cambios de brillo, etc.) que se aplicarán a las imágenes originales. Para CNN, la combinación de aumento de datos (Data Augmentation) para la clase minoritaria y el uso de ponderación de clases (Class Weights) suele ser la estrategia más robusta y recomendable, para reducir el desbalance de clases.
  3. Desarrollo y Entrenamiento del Modelo:
      - Diseño del modelo: Seleccionar la arquitectura de red neuronal o algoritmo de visión artificial adecuado para la tarea (por ejemplo, CNNs, R-CNN, YOLO, etc.).
      - Entrenamiento: Entrenar el modelo utilizando el conjunto de datos etiquetado. Esto a menudo implica el uso de técnicas como el aprendizaje por transferencia (transfer learning) si se dispone de datos limitados.
      - Validación y ajuste: Evaluar el rendimiento del modelo en un conjunto de datos de validación y ajustar los hiperparámetros para optimizar los resultados.
  4. Evaluación del Modelo:
      - Probar el modelo final con un conjunto de datos de prueba previamente no visto para medir su rendimiento real y verificar si cumple con los objetivos del proyecto.
      - Analizar métricas clave como precisión, recall, IoU (Intersección sobre Unión), y velocidad de inferencia.
  5. Implementación (Despliegue):
      - Integrar el modelo en el entorno de producción (servidor local, nube, dispositivo edge, etc.).
      - Asegurar la funcionalidad dentro del sistema de destino, que puede ser una aplicación de software, un dispositivo IoT o parte de un flujo de trabajo más amplio.
  6. Supervisión y Mantenimiento:
      - Monitorizar continuamente el rendimiento del modelo en tiempo real para detectar la desviación del modelo (model drift) o fallos.
      - Recopilar nuevos datos y reentrenar el modelo periódicamente para mantener su precisión y relevancia a lo largo del tiempo.
      - Realizar mantenimiento continuo y solucionar problemas operativos.

Estas fases son a menudo iterativas, donde el ajuste del modelo o la recopilación de datos adicionales pueden ser necesarios después de la evaluación o incluso del despliegue inicial.

---
```

#### Función para cargar los datasets desde disco duro local.
- Funciones de OpenCV (cv2) utilizadas :
  - cv2.imread carga un archivo de imagen desde una ubicación específica en el disco al programa, ***representándolo como una matriz NumPy***.
    - Abre el archivo de imagen en esa ruta.
    - ***Decodifica los datos de la imagen***.
    - ***Devuelve la imagen como una matriz NumPy multidimensional*** que se puede utilizar para su posterior procesamiento y análisis en Python. Por defecto, carga la imagen en formato de color BGR (azul, verde, rojo).
  - cv2.IMREAD_GRAYSCALE convierte, de ser el caso, las imágenes a escalas de grises.

  - cv2.resize(img_arr, (img_size, img_size)) se utiliza para cambiar el tamaño de una imagen a una dimensión fija específica. Acepta los siguientes argumentos:
    - img_arr
      - Esta es la imagen de origen, normalmente una matriz NumPy, que desea redimensionar.
    - (img_size, img_size)
      - Esta es la dimensión de salida deseada, proporcionada como una tupla de (**width, height**). Al usar el mismo valor tanto para el ancho como para la altura, la imagen se fuerza a tener forma cuadrada, lo que podría causar distorsión si la relación de aspecto original no es cuadrada.
        - La estructura del array de NumPy devuelto es un array de NumPy con la siguiente forma (shape): (**height, width**, channels).
        - channels será el mismo número de canales que la imagen original (img_arr).
        - Por lo tanto, la forma del array resultante (resized_arr.shape) será (img_size, img_size, C), donde C es el número de canales (por ejemplo, 3 para una imagen en color BGR, o 1 para una imagen en escala de grises).
        - **Es importante notar que OpenCV espera la tupla de dimensiones en el formato (width, height), a diferencia de la forma del array de NumPy que es (height, width, channels)**.
    - Usos clave
      - Estandarización para el aprendizaje automático : Un caso de uso común en visión por computadora es garantizar que todas las imágenes de entrada tengan las mismas dimensiones antes de ser introducidas en un modelo (por ejemplo, para la clasificación de imágenes).
      - Control del tamaño de la imagen : Permite aumentar (aumentar la escala) o disminuir (reducir la escala) el tamaño de una imagen en términos de su anchura y altura.
      - Procesamiento más rápido : Reducir el tamaño de la imagen puede acelerar el procesamiento en algunas aplicaciones.
    - Cómo funciona
      -La función interpola píxeles de la imagen original para crear la nueva imagen redimensionada. El método específico utilizado para esta interpolación se puede especificar con un parámetro de interpolation opcional; entre las opciones comunes se incluyen:
        - cv2.INTER_AREA Generalmente se prefiere para reducir/disminuir el tamaño de las imágenes.
        - cv2.INTER_LINEAR Método predeterminado, utilizado para redimensionar y hacer zoom en general.
        - cv2.INTER_CUBICO frece resultados más suaves y se utiliza normalmente para ampliar/aumentar la escala de imágenes.

```python
labels = ['PNEUMONIA', 'NORMAL']
img_size = 150
def cargar_datasets(data_dir):
    data = []
    for label in labels:
        path = os.path.join(data_dir, label) # Construir la ruta de archivo completa y correcta
        class_num = labels.index(label) # Encuentra la posición numérica en "labels" y la almacena en "class_num". "Pneumonia = 0". "Normal =1"
        for img in os.listdir(path): # Toma cada imagen en el directorio
            try:
                img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
                if img_arr is not None: # Verificar si la imagen fue cargada correctamente
                  resized_arr = cv2.resize(img_arr, (img_size, img_size))
                  data.append([resized_arr, class_num])
                else:
                    print(f"Warning: Could not read image {os.path.join(path, img)}")
            except Exception as e:
                print(e)
    return np.array(data, dtype=object) # Usar dtype=object si los elementos tienen formas diferentes, caso contrario arroja error.
```

No es estrictamente obligatorio tener los datasets en formato de numpy array para visión por computadora, pero es una práctica muy común y altamente recomendable.

**Razones:**
  - Eficiencia y rendimiento: Los numpy arrays están optimizados para operaciones numéricas y consumen menos memoria que las listas estándar de Python. Esto es crucial para manejar grandes volúmenes de datos de imágenes de manera eficiente.
  - Compatibilidad con librerías: La mayoría de los frameworks y librerías de visión por computadora y aprendizaje profundo (como **TensorFlow, PyTorch, Keras y OpenCV**) utilizan numpy arrays como el ***formato fundamental para representar y procesar imágenes***.
  - Operaciones vectorizadas: ***NumPy permite realizar operaciones matemáticas en matrices enteras sin la necesidad de bucles explícitos***, lo que hace que el código sea más rápido y conciso, facilitando tareas de preprocesamiento de imágenes.

**Alternativas y formatos comunes**
Si bien los numpy arrays son el formato de representación interna estándar durante el procesamiento, los datos de origen pueden estar almacenados en otros formatos:

  - Archivos de imagen: Los datos sin procesar se almacenan comúnmente como archivos JPEG, PNG, etc., y se cargan dinámicamente como numpy arrays (o tensores) durante el entrenamiento.
  - Formatos estandarizados: Los datasets populares utilizan formatos estandarizados como COCO o PASCAL VOC, que incluyen archivos de imagen junto con archivos de anotaciones (comúnmente en formato JSON o XML).
  - DataLoaders y ***TF.Dataset***: Para conjuntos de datos muy grandes que no caben en la memoria, se utilizan herramientas de streaming de datos como tf.data.Dataset de TensorFlow o los DataLoader de PyTorch. Estos cargan y procesan los datos por lotes (a menudo convirtiéndolos a ***tensores, que son similares a los numpy arrays pero optimizados para GPU***) desde el disco en tiempo real.

En resumen, aunque los datos brutos pueden almacenarse en diversos formatos, el ***numpy array es el lenguaje común para la manipulación y el cálculo numérico en el pipeline de visión por computadora.***

### Data Augmentation

Para evitar el sobre entrenamiento (overfitting) y el desbalance de clases, es necesario hacer mas grande nuestro dataset. Esto se puede lograr usando escalas de grises, rotaciones, traslaciones, desplazamientos horizontales o verticales y otros.

En nuestra modelación el autor original optó por lo siguiente:

   - Rotar algunas imagenes de entrenamiento 30 grados.
   - Aplicar zoom en un 20% a algunas imagenes.
   - Desplazar aleatoriamente las imágenes horizontalmente un 10% de su ancho.
   - Desplazar aleatoriamente las imágenes un 10% de su altura.
   - Voltear lasa imágenes en forma horizontal.
   - El "zca whitening" (Zero-Phase Component Analysis), se utiliza visión por computadora para transformar los datos de manera que su matriz de covarianza sea la matriz identidad. ***Esto elimina las correlaciones entre las características de los datos, hace que las variables sean independientes y estandariza sus varianzas***. El objetivo de la transformación ZCA es que los datos resultantes se mantengan lo más parecidos posible a los originales en términos de la correlación entre sus componentes.
  
Todo esto se hace utilizando la clase ImageDataGenerator de la librería Keras.

```python
datagen = ImageDataGenerator(
        featurewise_center=False,  # pone la media de los datos de entrada en cero en el dataset.
        samplewise_center=False,  # pone la media de cada dato en cero.
        featurewise_std_normalization=False,  # divide los datos de entrada para la desviación estandard del dataset
        samplewise_std_normalization=False,  # divide cada dato por su desviaci´pn estandar.
        zca_whitening=False,  # aplica ZCA whitening
        rotation_range = 30,  # rota imágenes.
        zoom_range = 0.2, # zoom imágenes
        width_shift_range=0.1,  # shift imágenes horizontalmente
        height_shift_range=0.1,  # shift imágenes verticalmente
        horizontal_flip = True,  # voltea imágenes
        vertical_flip=False)  # voltea imágenes


datagen.fit(x_train)
```

Es importante entender como funciona la clase ImageDataGenerator

El objeto ImageDataGenerator de Keras/TensorFlow, cuando se utiliza con métodos como flow(), flow_from_directory() o flow_from_dataframe(), **devuelve un generador (un iterador de Python)** que produce lotes (batches) de datos de imagen aumentados (y sus etiquetas correspondientes).

¿Qué es exactamente lo que devuelve?

  - Lotes de datos: El generador no devuelve todas las imágenes a la vez, sino en lotes (tamaño de lote especificado por el usuario) a medida que se solicitan, lo que es eficiente para la memoria.
  - Imágenes transformadas: Cada lote contiene las imágenes originales a las que se les han aplicado transformaciones aleatorias configuradas (como rotaciones, zoom, cambios de brillo, volteo horizontal, etc.) en tiempo real ("on-the-fly").
  - Etiquetas correspondientes: Junto con el lote de imágenes, el generador también devuelve las etiquetas asociadas a esas imágenes (por ejemplo, en formato one-hot encoded o sparse, dependiendo de la configuración class_mode).
  - Datos "frescos" en cada época: En cada época de entrenamiento, el generador produce variaciones ligeramente diferentes de las imágenes originales, lo que ayuda a prevenir el sobreajuste (overfitting) y mejora la generalización del modelo.

**En resumen**

***El ImageDataGenerator no altera el conjunto de datos original, sino que funciona como un suministrador dinámico de datos que "intercepta" las imágenes, las transforma aleatoriamente y las pasa al modelo en lotes durante el entrenamiento.***

#### Diseño de la red CNN

El diseño de la red, la selección del número de capas, strides, padding, dropout y otros hiperparámetros en un modelo de Red Neuronal Convolucional (CNN) es un proceso **empírico de prueba y error, no existe una fórmula matemática única**. La mejor configuración depende del problema específico, el conjunto de datos y el rendimiento deseado.

A continuación se presenta una guía basada en las mejores prácticas:

1. ***Transfer Learning*** (Recomendado para empezar)
  - La forma más efectiva de comenzar suele ser utilizar arquitecturas de CNN pre-entrenadas (como AlexNet, ResNet, VGG, etc.) mediante transfer learning. Estas redes ya han demostrado un excelente rendimiento en tareas de visión artificial y se pueden ajustar sus capas finales al problema específico.
2. Principios de Diseño y Ajuste de Hiperparámetros

  - Si se opta por diseñar una arquitectura propia desde cero, se deben considerar los siguientes puntos:

    - Número de Capas Convolucionales
      - Complejidad del Problema: Para problemas simples (ej. clasificación de dígitos MNIST), pocas capas (2-3) son suficientes. Para problemas complejos (ej. detección de objetos en imágenes de alta resolución), se necesitan redes más profundas para extraer características jerárquicas.
      - Extracción de Características: Las primeras capas aprenden características básicas (bordes, texturas), mientras que las capas más profundas aprenden características más abstractas (formas, objetos).
      - Rendimiento: Empezar con una arquitectura simple; y añadir capas si el modelo no alcanza un rendimiento adecuado (subajuste o underfitting).
    - Strides (Pasos)
      - Reducción de Dimensionalidad: Un stride mayor a 1 (comúnmente 2) reduce el tamaño espacial del mapa de características. Esto se suele usar en lugar o en combinación con las capas de max-pooling para hacer la red más eficiente computacionalmente.
      - Compromiso: Strides grandes pueden hacer que se pierda información espacial valiosa, mientras que strides pequeños mantienen más detalles a costa de un mayor costo computacional.
    - Padding (Relleno)
      - Control del Tamaño: El padding (generalmente con ceros) se usa para asegurar que el tamaño espacial del mapa de características no disminuya demasiado rápido con cada capa convolucional, especialmente en los bordes de la imagen.
      - "Same" vs "Valid": "Same" padding mantiene el mismo tamaño de salida que de entrada, mientras que "valid" padding reduce el tamaño. "Same" es común para mantener la información de los bordes.
    - Dropout (Regularización)
      - Prevención de Overfitting: El dropout es una técnica de regularización crucial. Consiste en desactivar aleatoriamente un porcentaje de neuronas durante el entrenamiento, lo que evita que el modelo dependa en exceso de neuronas o características específicas, haciéndolo más robusto.
      - Ajuste: El valor de dropout (ej. 0.2 a 0.5) se ajusta si el modelo muestra signos de sobreajuste (overfitting).
3. Métodos de Optimización

  - El ajuste de hiperparámetros se realiza normalmente mediante:
    - Prueba y Error Manual: Un enfoque intuitivo y común para empezar.
    - Búsqueda en Cuadrícula (Grid Search): Probar sistemáticamente todas las combinaciones posibles de un conjunto predefinido de hiperparámetros.
    - Búsqueda Aleatoria (Random Search): Probar combinaciones aleatorias, a menudo más eficiente que la búsqueda en cuadrícula.
    - Optimización Bayesiana/Algoritmos Genéticos: Métodos más avanzados y automatizados que buscan la mejor combinación de manera más inteligente.

En resumen, la mejor manera es ***empezar con una arquitectura probada*** o una configuración de referencia, y luego ***usar un conjunto de validación*** para evaluar el rendimiento ***mientras se ajustan los hiperparámetros de forma iterativa***.

#### Intentemos nuestro propo diseño

```python
model = Sequential()
model.add(Conv2D(32 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu' , input_shape = (150,150,1)))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Conv2D(64 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(Dropout(0.1))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Conv2D(64 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Conv2D(128 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Conv2D(256 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Flatten())
model.add(Dense(units = 128 , activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units = 1 , activation = 'sigmoid'))
model.compile(optimizer = "rmsprop" , loss = 'binary_crossentropy' , metrics = ['accuracy'])
model.summary()
```

La matriz de confusión es una tabla que visualiza el rendimiento del modelo al desglosar los tipos de aciertos y errores que ha cometido en sus predicciones en comparación con los valores reales.

Componentes Clave de la Matriz de Confusión

Para un problema de clasificación binaria, la matriz usualmente tiene dos filas (clase real) y dos columnas (clase predicha), dividida en cuatro cuadrantes:

  - Verdaderos Positivos (VP): El modelo predijo la clase positiva y la clase real también era positiva. Son aciertos del modelo para la clase de interés.
  - Verdaderos Negativos (VN): El modelo predijo la clase negativa y la clase real también era negativa. Son aciertos del modelo para la clase que no es de interés.
  - Falsos Positivos (FP): El modelo predijo la clase positiva, pero la clase real era negativa. Es un error de tipo I (el modelo se equivocó al clasificar algo como positivo).
  - Falsos Negativos (FN): El modelo predijo la clase negativa, pero la clase real era positiva. Es un error de tipo II (el modelo se equivocó al clasificar algo como negativo).

Interpretación Visual y Métrica

  - Diagonal Principal (VP + VN): Los valores en la diagonal principal representan las predicciones correctas (aciertos) del modelo. Un modelo perfecto tendría todos sus datos en esta diagonal.
  - Fuera de la Diagonal (FP + FN): Los valores fuera de la diagonal principal representan las predicciones incorrectas (errores o confusiones) del modelo.

Uso para Calcular Métricas de Rendimiento

La interpretación va más allá de un simple vistazo a la matriz. Los valores individuales se utilizan para calcular métricas de evaluación más específicas que la simple exactitud (accuracy):

  - Exactitud (Accuracy): Proporción total de predicciones correctas (VP + VN) / Total.
  - Precisión (Precision): De todas las veces que el modelo predijo positivo, cuántas veces acertó (VP / (VP + FP)). Relevante cuando el costo de un Falso Positivo es alto.
  - Sensibilidad (Recall o Exhaustividad): De todos los casos positivos reales, cuántos detectó correctamente (VP / (VP + FN)). Relevante cuando el costo de un Falso Negativo es alto (ej. detección de enfermedades).
  - Puntuación F1 (F1-Score): Una media armónica de precisión y sensibilidad, útil cuando se busca un equilibrio entre ambas.
   
En resumen, la matriz de confusión proporciona un desglose detallado del rendimiento de la CNN, permitiendo entender no solo cuántas veces acertó, sino dónde y cómo se equivocó, lo cual es fundamental para mejorar el modelo.

### Cómo interpretar los gráficos de tendencias.

Representar gráficamente las métricas de entrenamiento y validación frente al número de épocas es la práctica habitual.
1. Entrenamiento ideal
  - Loss: Tanto la de entrenamiento como la de validación disminuyen y se mantienen cercanas entre sí.
  - Accuracy: Tanto la del entrenamiento como la de la validación aumentan y se mantienen cercanas entre sí.
  - Interpretación: ***El modelo aprende eficazmente y generaliza bien a nuevos datos. Podría entrenarse durante algunas épocas más si la tendencia sigue mejorando***.
2. Sobreajuste (overfitting)
  - Loss: La de entrenamiento continúa disminuyendo, pero la de validación comienza a aumentar después de cierto punto.
  - Accuracy: La del entrenamiento continúa aumentando (a menudo acercándose al 100%), pero la de la validación se estanca o comienza a disminuir .
  - Interpretación: ***El modelo ha comenzado a aprender ruido y detalles específicos de los datos de entrenamiento que no se aplican a datos no vistos. Esto indica que el entrenamiento debería haberse detenido antes (lo que se conoce como parada temprana) o que se necesitan técnicas de regularización (por ejemplo, dropout, aumento de datos, etc.).***
3. Ajuste insuficiente (underfitting)
  - Loss: Tanto la de entrenamiento como la de validación son altas y se estabilizan (aplanan) rápidamente, o la disminución es muy lenta.
  - Accuracy: Tanto la del entrenamiento como la de la validación son bajas y se estancan.
  - Interpretación: ***El modelo es demasiado simple (p. ej., no tiene suficientes capas ni parámetros) para capturar los patrones subyacentes en los datos. Es posible que necesite una arquitectura de modelo más compleja o más características.***

4. Alta varianza (gran brecha)
  - Pérdida/Precisión: Ambas métricas muestran las tendencias esperadas (pérdida a la baja, precisión al alza), pero existe una gran brecha entre las curvas de entrenamiento y validación, incluso si el rendimiento de la validación sigue mejorando.
  - Interpretación: ***El modelo aprende muy bien con los datos de entrenamiento, pero tiene dificultades para generalizar a los datos de validación***. Esto suele ser un síntoma de sobreajuste, y técnicas como la regularización o la adquisición de más datos de entrenamiento podrían ayudar.

El desarrollo de modelos CNN implica un grado de ***experimentación y refinamiento sistemático***, más que pura "prueba y error" aleatorio. La decisión sobre qué modelo es el "mejor" se basa en una evaluación rigurosa utilizando métricas de rendimiento, y considerando criterios prácticos como los recursos computacionales y los requisitos del problema.

**Criterios Clave para Decidir "Qué es lo Mejor"**

La selección del mejor modelo se basa en un equilibrio de los siguientes factores:

1. Métricas de Rendimiento en el ***Conjunto de Validación***:
- Accuracy, Recall, F1 Score, mAP (mean Average Precision):
    - Estas métricas objetivas miden qué tan bien generaliza el modelo a datos no vistos durante el entrenamiento. ***El mejor modelo es el que logra el rendimiento más alto en el conjunto de validación.***
- Pérdida (Loss):
    - Se monitoriza la ***pérdida tanto en el conjunto de entrenamiento como en el de validación***. Un ***modelo óptimo*** tendrá una ***pérdida de validación baja y comparable a la pérdida de entrenamiento***, evitando el sobreajuste (overfitting) o subajuste (underfitting).
2. Generalización a Datos de Prueba:
    - Una vez seleccionado el mejor modelo usando el conjunto de validación, ***se evalúa en un conjunto de prueba completamente independiente*** para obtener una estimación imparcial de su rendimiento en el mundo real. ***El resultado en este conjunto es el indicador final de su calidad***.
3. Eficiencia Computacional y Restricciones de Despliegue:
    - Un modelo con una ***precisión ligeramente inferior*** pero que es ***mucho más rápido de entrenar y ejecutar*** (inferencia) ***puede ser preferible*** si los recursos computacionales son limitados o si se necesita un despliegue en tiempo real (por ejemplo, en dispositivos móviles).
    - El tamaño del modelo (***número de parámetros***) influye directamente en estos aspectos.
4. Requisitos Específicos del Problema:
    - ***Algunas aplicaciones pueden priorizar la interpretabilidad, mientras que otras exigen la máxima precisión posible, sin importar la complejidad del modelo. La elección debe alinearse con estos objetivos***.

Enfoque Sistemático (Más Allá de la Prueba y Error)

- Aunque la experimentación es inherente, el proceso puede ser sistemático:

    - Comenzar con arquitecturas establecidas:
        - Utilizar modelos preentrenados (transfer learning) como VGG, ResNet, Inception o MobileNet, que han demostrado un rendimiento sólido en tareas similares, es un excelente punto de partida.
    - Ajuste de hiperparámetros:
        - En lugar de probar configuraciones al azar, se pueden utilizar métodos como la búsqueda en cuadrícula (Grid Search), la búsqueda aleatoria (Random Search) u optimización bayesiana para explorar el espacio de hiperparámetros de manera más eficiente.
    - Herramientas de búsqueda automática (NAS):
        - Métodos avanzados como la ***Búsqueda de Arquitectura Neuronal (Neural Architecture Search - NAS)*** utilizan algoritmos (como el aprendizaje por refuerzo o algoritmos genéticos) para diseñar automáticamente la arquitectura óptima, reduciendo la intervención manual.
    
***Se decide que un modelo es "el mejor" cuando ha demostrado el mejor equilibrio entre rendimiento (precisión) y eficiencia para los objetivos específicos del proyecto, validado en un conjunto de datos de prueba independiente.***

### Hyperparameter tunning

***¿Se deben seleccionar los hiper parámetros, incluyendo el número de capas convolucionales primero de una forma manual y luego entrenar el modelo, ejecutar GridSearchCv para seleccionar los mejores parámetros y finalmente entrenar al modelo final?.***

**No, lo que la pregunta sugiere no es lo estandar, ni la única práctica; existen métodos mas eficientes e integrados.**

La práctica estándard consiste en utilizar métodos de búsqueda sistemáticos, como GridSearch o mas comunmente in Deep Learning, ***RandomSearch u optimización Bayesiana*** para explorar un rango de parámetros predefinidos.

1. Definir un espacio de búsqueda:
  - Primero se define un grupo de posibles valores para todos los hiper parámetros que se desea optimizar. Esto puede incluir:
    - Network Architecture: El número de capas convolucionales, el  número de filtros por capa, el tamaño del filtro (kernel), el numero de capas densas, etc.
    - Parametros de optimización: Learning rate, batch size, optimizador, momentum, etc.
    - Regularización: Dropout rate, L1/L2 regularization strengths
  - Se puede usar la experiencia aquirida o la intuición para setear un rago de valores razonable para la búsqueda, pero no un único valor fijo que requiera un preentrenamiento separado.
2. Ejecutar GridSearch con Cross-Validation:
  - Grid Search CV automáticamente entrena y evalúa el modelo para cada combinación en el espacio definido.
  - Esta evaluación se hace usando el ***set de datos de validación*** o k-fold cross-validation para lograr una estimación robusta del  modelo y prevenir "overfitting".
  - **Importante** Si se usara el set de datos de prueba para ajustar los hiperparámetros, se estaría, en efecto, "entrenando" indirectamente al modelo (y sus hiperparámetros) para que funcione bien en los datos de prueba. Esto resultaría en una métrica de rendimiento sesgada y poco realista sobre cómo se desempeñaría el modelo en el mundo real. El ***set de datos de prueba debe ser la prueba final*** y definitiva de la capacidad de generalización del modelo.
  - Los modelos se entrenan desde cero con cada nuevo set de hiper parámetros.
3. Seleccionar los mejores parámetros:
  - Despues de que la búsqueda se ha completado, se selecciona la combinación de hiper parámetros que  nos dan el mejor desempeño.( highest validation accuracy o lowest validation loss).
4. Entrenar el modelo final:
  - El modelo final es entonces entrenado usando todo el data set de entrenamento (combinando los data sets originales de entrenamiento y validación, de ser aplicable) conlos parámetros óptimos encontrados durante el proceso de búsqueda (search).

***¿Porque el flujo descrito en la pregunta no es necesario?***

- El propósito de GridSearch es evaluar sistemáticamente diferentes combinaciones de parámetros, por lo que no es necesario primero entrenar manualmente al modelo.
- La arquitectura de la red puede ser un hiper parámetro:
  - El número de capas convolucionales puede ser incluido como un parámetro  ejemplo: num_layers: [1, 2, 3]).
- Eficiencia: La selección manual, entrenamiento y luego optimizar los arámetros  con Grid Search es ineficiente. Los métodos automatizados gilitan el proceso. Para Deep Learning, random search o Bayesian optimization son con frecuencia preferidos sobre un GridSEarch exhaustivo debido principalmente a los altos costos computacionales.

En resumen, la mejor práctica es integrar la selección de parámetros en un proceso sistemático de optimización desde el comienzo. Se define el rango de valores potenciales y se deja que el algoritmo de optimización y la validación cruzada encuentren la configuración óptima.
